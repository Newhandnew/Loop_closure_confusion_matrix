```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
```

```python
fig.set_size_inches(28, 12)
```

```python
dataGroundTruth = np.loadtxt("NewCollegeTextFormat.txt", delimiter = ",")
```

```python
sns.heatmap(dataGroundTruth, ax = ax1, xticklabels=100, yticklabels=100, cbar_kws={'label': 'groundtruth'})
```

```python
ax1.set_title('Ground truth', fontsize=24)
```

```python
ax1.set_xlabel('image number', fontsize=24)
```

```python
ax1.set_ylabel('image number', fontsize=24)
```

```python
dataBow = np.loadtxt("surf100000_college.txt")
```

```python
# get lower triangle and -1 means below Diagonal 1
dataBow = np.tril(dataBow, -1)
```

```python
sns.heatmap(dataBow, ax = ax2, vmax=0.2, xticklabels=100, yticklabels=100, cmap="Greens", cbar_kws={'label': 'score'})
```

```python
ax2.set_title('Surf BOW', fontsize=24)
```

```python
ax2.set_xlabel('image number', fontsize=24)
```

```python
ax2.set_ylabel('image number', fontsize=24)
```

```python
plt.savefig('matrix.png')
```

```python
plt.show()
```
